```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x, mean, var):
    return 1/np.sqrt(2*np.pi*var**2)*np.exp(-(x-mean)**2/(2*var**2))

x = np.linspace(0, 100, 1000)
E_stokes = gauss(x, 75, 1)
E_astokes = gauss(x, 25, 1)
E_rayleigh = 10*gauss(x, 50, 1)
spectrum =  E_stokes+E_astokes+E_rayleigh
xticks = np.arange(0, 112.5, 12.5)
plt.rcParams['font.size'] = 14
plt.plot(x, spectrum, color='blue')
xticklabels = ['', '', 'E-$\Delta$E', '', 'E', '', 'E+$\Delta$E', '', '']
plt.text(52, 1.7, 'Rayleigh')
plt.text(20, 0.55, 'Stokes')
plt.text(67, 0.55, 'anti-Stokes')
plt.xticks(xticks, labels=xticklabels)
plt.yticks(plt.yticks()[0], labels='')
plt.ylabel('Intensity')
plt.xlabel('Scattered photon energy')
plt.grid()
plt.savefig(r'.\Scatter.pdf')
plt.show()
```
